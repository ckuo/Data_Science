## Performing PCA

```python
# scree_plot

def scree_plot(pca_object, export = False):
    # building a scree plot

    # setting plot size
    fig, ax = plt.subplots(figsize=(10, 8))
    features = range(pca_object.n_components_)


    # developing a scree plot
    plt.plot(features,
             pca_object.explained_variance_ratio_,
             linewidth = 2,
             marker = 'o',
             markersize = 10,
             markeredgecolor = 'black',
             markerfacecolor = 'grey')


    # setting more plot options
    plt.title('Scree Plot')
    plt.xlabel('PCA feature')
    plt.ylabel('Explained Variance')
    
    # Example cutoff at the 4th component (which is index 3)
    plt.axvline(x=3, linestyle=':', color='red')
    
    plt.xticks(features)

    # forcing the plot to display the vertical line
    plt.draw()

    # displaying the plot
    plt.show()



########################################
# unsupervised_scaler
########################################
def scaler(df):
    """
    Standardizes a dataset (mean = 0, variance = 1). Returns a new DataFrame.
    Requires sklearn.preprocessing.StandardScaler()
    
    PARAMETERS
    ----------
    df     | DataFrame to be used for scaling
    """

    # INSTANTIATING a StandardScaler() object
    scaler = StandardScaler(copy = True)


    # FITTING the scaler with the data
    scaler.fit(df)


    # TRANSFORMING our data after fit
    x_scaled = scaler.transform(df)

    
    # converting scaled data into a DataFrame
    new_df = pd.DataFrame(x_scaled)


    # reattaching column names
    new_df.columns = list(df.columns)
    
    return new_df
########################################
## visual_cm
########################################
def visual_cm(true_y, pred_y, labels = None):
    """
Creates a visualization of a confusion matrix.

PARAMETERS
----------
true_y : true values for the response variable
pred_y : predicted values for the response variable
labels : , default None
    """
    # visualizing the confusion matrix

    # setting labels
    lbls = labels
    

    # declaring a confusion matrix object
    cm = confusion_matrix(y_true = true_y,
                          y_pred = pred_y)


    # heatmap
    sns.heatmap(cm,
                annot       = True,
                xticklabels = lbls,
                yticklabels = lbls,
                cmap        = 'Blues',
                fmt         = 'g')


    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix of the Classifier')
    plt.show()
```

```python
# dropping other colums
facebook_type = facebook_df.drop(['status_id', 'status_type', 'time_published'],
                                      axis = 1)


# applying the unsupervised_scaler function
facebook_type_scaled = scaler(df = facebook_type)


# checking pre- and post-scaling variance
print(np.var(facebook_type), '\n\n')
print(np.var(facebook_type_scaled))
```

```python
# grouping data based on Ward distance
standard_mergings_ward = linkage(y = facebook_type_scaled,
                                 method = 'ward',
                                 optimal_ordering = True)


# setting plot size
fig, ax = plt.subplots(figsize=(12, 12))

# developing a dendrogram
dendrogram(Z = standard_mergings_ward,
           leaf_rotation  = 90       ,
           leaf_font_size = 6        )


# rendering the plot
plt.show()
```

```python
# INSTANTIATING a PCA object with no limit to principal components
pca = PCA(n_components = None,
          random_state = 702)


# FITTING and TRANSFORMING the data
facebook_pca = pca.fit_transform(facebook_type_scaled)


# calling the scree_plot function
scree_plot(pca_object = pca)
```

This scree plot displays the variance explained by each principal component in the PCA analysis. The y-axis shows the variance explained, and the x-axis lists the principal components. The first point (PC0) is the highest, explaining the most variance. The second point (PC1) shows a significant drop but still accounts for a good amount of variance. After PC2, marked by the dotted red line, there's an 'elbow' in the plot, where the line begins to flatten out, indicating that additional components explain progressively less variance. This leveling off suggests that PC0 through PC2 are the most informative for understanding the structure in the data, and further components contribute less to the total variance explained. 

### The first dot (PC 0: 0.45) 
For PC0, which represents the first principal component, its high value of approximately 45% indicates that it explains a significant portion of the variance in the dataset, suggesting that it captures the most prominent patterns or trends within the data. This makes PC0 crucial for understanding the fundamental structure of the dataset and identifying its primary underlying features. Additionally, PC0's dominance in variance explained implies that it may correspond to a key dimension along which the data points vary the most. Therefore, analyzing the relationship between variables and PC0 can offer valuable insights into the primary factors driving variation in the dataset.


### The second dot (PC 1: 0.125) 
Moving to PC1, which accounts for around 12.5% of the variance, it represents a substantial but comparatively smaller portion of the dataset's variability. Despite this, PC1 still captures meaningful patterns that differ from those captured by PC0, providing additional insights into the dataset's structure. By examining the loading scores of variables on PC1, one can identify the specific aspects of the data that contribute most to this component's variance. This analysis can help uncover secondary patterns or trends that are distinct from those identified by PC0, enriching the overall understanding of the dataset.


### The third dot (PC 2: 0.12) 
Finally, PC2, with an explained variance of approximately 12%, contributes further to understanding the dataset's structure by capturing additional patterns not fully represented by PC0 and PC1. While PC2 explains less variance than PC0 and PC1, its insights are still valuable for gaining a comprehensive understanding of the dataset's underlying characteristics. By exploring the relationship between variables and PC2, I can uncover unique features or relationships that contribute to the dataset's ingormation. Thus, collectively analyzing PC0, PC1, and PC2 allows for a more understanding of the dataset's variability and underlying structure.

Based on these metrics, the best model is typically the one with the highest Pseudo R-squared (indicating a better fit), the lowest AIC and BIC (indicating a more parsimonious model), and the smallest LLR p-value (indicating the model is a significant improvement over the null model).

Model 1 and Model 2 are quite similar in terms of Pseudo R-squared, but Model 2 has a slightly lower AIC and BIC, which suggests it is a more efficient model.
Model 3 has the lowest Pseudo R-squared and the highest AIC and BIC, indicating it might not fit the data as well as the other models.

Therefore, Model 2 appears to be the best model among the three, given its balance of model fit. It fits the data nearly as well as Model 1 (as indicated by the Pseudo R-squared) but does better (as indicated by the lower AIC and BIC), and it has the highest LLR p-value, which suggests a very low likelihood that the model's relationship with the dependent variable is due to chance.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import pandas as pd


# Perform train-test split using the selected features for X
X = facebook_df[model2_features]
y = facebook_df['status_type']

x_train, x_test, y_train, y_test = train_test_split(
    X, 
    y,
    test_size=0.25,
    random_state=219,
    stratify=y)  # Make sure to stratify by y to preserve class balance

# Instantiate the logistic regression model
logreg = LogisticRegression(solver='lbfgs', max_iter=200, random_state=219)

# Fit the model on training data
logreg_fit = logreg.fit(x_train, y_train)

# Predict and evaluate the model
y_pred_train = logreg_fit.predict(x_train)
y_pred_test = logreg_fit.predict(x_test)

train_accuracy = accuracy_score(y_train, y_pred_train)
test_accuracy = accuracy_score(y_test, y_pred_test)
train_test_gap = abs(train_accuracy - test_accuracy)

# Print model evaluation results
print(f"Training ACCURACY: {train_accuracy:.4f}")
print(f"Testing ACCURACY: {test_accuracy:.4f}")
print(f"Train-Test Gap: {train_test_gap:.4f}")
```

```python
from sklearn.metrics import roc_auc_score, precision_score, recall_score

# Get the predicted probabilities for the positive class
y_score = logreg_fit.predict_proba(x_test)[:, 1]  # Probabilities for the positive class

# Calculate AUC
auc = round(roc_auc_score(y_true=y_test, y_score=y_score), ndigits=4)

# Calculate precision and recall using the predictions
precision = round(precision_score(y_true=y_test, y_pred=y_pred_test), ndigits=4)
recall = round(recall_score(y_true=y_test, y_pred=y_pred_test), ndigits=4)

# Print the results
print(f"AUC: {auc}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
```
