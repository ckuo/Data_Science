```python
import time
import random

character_lst = [
    "Indiana Jones - The Daring Archaeologist",
    "Marion Ravenwood - The Resourceful Ally",
    "Henry Jones Sr. - The Scholarly Father",
    "Sallah - The Loyal Excavator",
    "Mutt Williams - The Rebellious Adventurer",
    "Irina Spalko - The Formidable Antagonist"
]


stage_lst = ['temple', 'jungle', 'river', 'cave']



def welcome():
    banner = """
        ⠀⠀⠀⠀⠀⠀⠀⢀⣤⣴⣶⣶⣿⣿⣿⣿⣿⣷⣶⣶⣦⣤⡀⠀⠀⠀⠀⠀⠀⠀
        ⠀⠀⠀⠀⠀⠀⠀⢸⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⡇⠀⠀⠀⠀⠀⠀⠀
        ⠀⠀⠀⠀⠀⠀⠀⣾⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣧⠀⠀⠀⠀⠀⠀⠀
        ⠀⠀⠀⠀⠀⠀⠰⠛⠋⠉⠉⠉⠀⠀⠀⠀⠀⠀⠉⠉⠉⠙⠛⠄⠀⠀⠀⠀⠀⠀
        ⠀⠀⠀⠀⠀⠀⢀⣀⣀⣀⣤⣤⣤⣤⣤⣤⣤⣤⣤⣤⣄⣀⣀⡀⠀⠀⠀⠀⠀⠀
        ⣶⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣷⣦
        ⠹⣷⡉⠛⠻⢿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⢿⣿⡿⠛⠋⣡⣾⠃
        ⠀⠈⠻⣶⣀⣸⣿⡇⠀⢀⣭⣭⣭⠉⠉⠉⠉⣭⣭⣤⣤⡄⢸⣿⣇⣀⣶⠟⠁⠀
        ⠀⠀⠀⠈⣿⡏⠀⠀⠐⠉⢥⣤⠀⠀⠀⠀⠀⠀⣴⡤⠀⠀⠀⠀⢹⣿⠁⠀⠀⠀
        ⠀⠀⠀⠀⢹⣧⡀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⣸⡟⠀⠀⠀⠀
        ⠀⠀⠀⠀⠀⠉⣿⡇⠀⠶⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠶⠀⢸⡿⠉⠀⠀⠀⠀⠀
        ⠀⠀⠀⠀⠀⠀⠸⣷⡆⠀⠀⠶⢀⣀⣀⣀⣀⡀⠶⠀⠀⢰⣾⠇⠀⠀⠀⠀⠀⠀
        ⠀⠀⠀⠀⠀⠀⠀⠘⢷⣟⠀⠀⠈⠉⠉⠉⠉⠁⠀⠀⣻⡾⠃⠀⠀⠀⠀⠀⠀⠀
        ⠀⠀⠀⠀⠀⠀⠀⠀⠈⠻⣷⣟⠀⡀⠘⠃⢀⠀⣻⣾⠟⠁⠀⠀⠀⠀⠀⠀⠀⠀
        ⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀⠈⠙⠿⣷⣤⣤⣾⠿⠛⠁⠀⠀⠀⠀⠀⠀⠀⠀⠀⠀
    """
    print(banner)
    """
    Prints a welcome message and explains the rules and objective of the game inspired by the Indiana Jones series.
    """
    print("Welcome to the Indiana Jones Adventure Game!")
    print("\nEmbark on a journey filled with ancient mysteries and daring exploits, reminiscent of the legendary adventures of Indiana Jones.")
    print("Your mission is to navigate through a series of challenges, including exploring lost temples, deciphering ancient riddles, evading treacherous traps, and uncovering hidden artifacts.")
    print("\nRules:")
    print("1. In each stage, you will face decisions that test your wit and courage, much like Indiana Jones himself. Choose wisely to advance.")
    print("2. You have three attempts to succeed in each challenge. A bad choice might lead to perilous outcomes.")
    print("3. Complete your quest within 180 seconds to claim the ancient treasure and achieve glory.")
    print("\nAre you ready to step into the shoes of the great adventurer? Let the quest begin!\n")




def choose_character():
    """
    Allows the player to choose a character from the list and returns the character's name.
    """
    print("Choose your character:")
    for i, character in enumerate(character_lst, 1):
        print(f"{i}. {character}")

    while True:
        try:
            choice = int(input(prompt="Enter the number of your character: "))
            if 1 <= choice <= len(character_lst):
                # Extracting just the name of the character
                return character_lst[choice - 1]
            else:
                print("Invalid choice. Please select a number from the list.")
        except ValueError:
            print("Invalid input. Please enter a number.")


def win(p_name):
    """
    Show the win message and ask user if they want to replay.
    """
    banner = """
         __      __.__
/  \    /  \__| ____   ____   ___________
\   \/\/   /  |/    \ /    \_/ __ \_  __ \
 \        /|  |   |  \   |  \  ___/|  | \/
  \__/\  / |__|___|  /___|  /\___  >__|
       \/          \/     \/     \/
    """

    print(banner)
    choice = input(prompt=f"\nCongratulations! You won as {p_name}! Would you like to play again? (yes/no): ")
    if choice.strip().casefold() == "yes":
        run_game()
    else:
        print("Thanks for playing! Goodbye.")



def fail(p_name):
    """
    Show the loosing message and ask user if they want to replay.
    """
    banner = """
        /$$
        | $$
        | $$        /$$$$$$   /$$$$$$   /$$$$$$$  /$$$$$$
        | $$       /$$__  $$ /$$__  $$ /$$_____/ /$$__  $$
        | $$      | $$  \ $$| $$  \ $$|  $$$$$$ | $$$$$$$$
        | $$      | $$  | $$| $$  | $$ \____  $$| $$_____/
        | $$$$$$$$|  $$$$$$/|  $$$$$$/ /$$$$$$$/|  $$$$$$$
        |________/ \______/  \______/ |_______/  \_______/


    """
    print(banner)
    choice = input(prompt=f"\nYou lose as {p_name}. Would you like to try again? (yes/no): ")
    if choice.strip().casefold() == "yes":
        run_game()
    else:
        print("Better luck next time!")



def try_stage(stage_function):
    """
    User gets 3 attempts at each stage
    Returns True if stage is successfully completed
    else it returns Fasle
    """
    attempts = 3
    while attempts > 0:
            # stage_function is any one of the challenges i.e temple, jungle, river or cave
        success = stage_function()
        if success:
            return True
        
        attempts -= 1
        print(f"You have made a bad choice. You have {attempts} attempts remaining.")
    return False


def ask_for_choice(prompt):
    """
    Ask user to choose from 2 options.
    Exception for if user enters non integer input.
    returns the choice, user has made.
    """
    choice = None
    is_valid_choice = False
    while not is_valid_choice:
        try:
            choice = int(input(prompt=prompt))
        except ValueError:
            print("Invalid type of choice.")
            
        if choice == 1 or choice == 2:
            is_valid_choice = True
        else:
            print("Not a valid choice! Please try again.")
            
    return choice

# Temple Stage
def temple():
    """
    Option 1: Coin Flip.
    Option 2: Solve a maths multiplication question.
        Exception if user answers with anything except an integer answer
        Player loses if answers takes more than 5 seconds.
    """
    print("\n--- Temple ---")
    print("You enter a mysterious temple with two doors.")
    choice = ask_for_choice("1. Go through the left door.\n2. Go through the right door.\nEnter your choice (1 or 2): ")

    if choice == 1:
        print("You encounter a guardian who challenges you to guess a coin flip.")
        coin_flip_guess = None
        is_valid_option = False
        while not is_valid_option:
            coin_flip_guess = input(prompt="Guess 'heads' or 'tails': ").lower()
            if coin_flip_guess in ['heads', 'tails']:
                is_valid_option = True
            else:
                print("Not a valid choice! Please try again.")

        coin_flip_result = random.choice(['heads', 'tails'])
        if coin_flip_guess == coin_flip_result:
            print(f"You guessed right! The coin shows {coin_flip_result}.")
            print("You found a map.")
            return True
        else:
            print(f"You guessed wrong. The coin shows {coin_flip_result}.")
            return False
        
    elif choice == 2:
        print("You encounter a wise sage who asks you a math question.")
        answer = random.randint(1, 5) * random.randint(1, 5)
        print(f"What is {answer} multiplied by 5? You have 5 seconds to answer.")

        start_time = time.time()
        
        try:
            user_answer = int(input())
        except ValueError:
            print("Invalid answer!")
            return False
        
        end_time = time.time()

        if end_time - start_time > 5:
            print("Too slow! You didn't answer in time.")
            return False
        elif int(user_answer) == answer * 5:
            print("Correct answer!")
        else:
            print("Wrong answer!")
            return False
    else:
        return True


def jungle():
    """
    Option 1. Bad poisonous plants.
    Option 2. Good choice for navigating the jungle.
    """
    print("\n--- Jungle ---")
    print("You enter a dense jungle. The path ahead splits in several directions.")
    print("To navigate through, you must choose the correct path at each turn.")

    # First Choice
    choice1 = ask_for_choice("1. Take the path with strange flowers.\n2. Follow the sound of running water.\nEnter your choice (1/2): ")
    
    if choice1 == 1:
        print("You encounter a dead end with poisonous plants. You have to turn back.")
        return False
    elif choice1 == 2:
        print("The sound leads you to a clearing with two more paths.")

        # Second Choice
        choice2 = ask_for_choice("1. Climb a tree to survey the area.\n2. Continue on the ground path.\nEnter your choice (1/2): ")
        
        if choice2 == 1:
            print("From the tree, you spot a hidden path leading out of the jungle.")
            return True
        elif choice2 == 2:
            print("You wander aimlessly and lose precious time.")
            return False
    else:
        return False
    
    return False


# River Stage
def river():
    """
    Option 1: Cross the bridge by answering a riddle.
    Option 2: Use raft to cross the river.
    """
    print("\n--- River ---")
    print("You reach a wide river with multiple ways to cross.")
    print("Choosing the right method is key to moving forward.")

    # First Choice
    choice1 = ask_for_choice("1. Look for a bridge.\n2. Attempt to build a raft.\nEnter your choice (1/2): ")
    if choice1 == 1:
        print("You find a bridge, but it's guarded by a riddle.")

        # Riddle
        riddle_answer = input("I speak without a mouth and hear without ears. I have no body, but I come alive with wind. What am I? ")
        if riddle_answer.casefold() == "echo":
            print("Correct! The guardian lets you pass.")
            return True
        else:
            print("Incorrect. You cannot cross the bridge.")
            return False
    elif choice1 == 2:
        print("You successfully build a raft but it takes longer than expected.")
        return True
    else:
        return False
    
    return False

# Cave Stage
def cave():
    """
    Option 1: Hidden door and rock, paper and scissors.
    Option 2: Getting lost.
    """
    print("\n--- Cave ---")
    print("You enter a dark cave with mysterious inscriptions on the walls.")
    print("Deciphering them might lead to the treasure.")

    # First Choice
    choice1 = ask_for_choice("1. Examine the inscriptions.\n2. Ignore them and proceed forward.\nEnter your choice (1/2): ")
    if choice1 == 1:
        print("The inscriptions reveal a hidden door. You proceed through it.")

        # Rock-Paper-Scissors Game
        print("A guardian challenges you to a game of Rock-Paper-Scissors.")
        user_choice = None
        is_valid_option = False
        while not is_valid_option:
            user_choice = input(prompt="Choose 'rock', 'paper', or 'scissors': ").lower()
            if user_choice in ['rock', 'paper', 'scissors']:
                is_valid_option = True
            else:
                print("Not a valid choice! Please try again.")
            
        guardian_choice = random.choice(['rock', 'paper', 'scissors'])
        print(f"The guardian chooses {guardian_choice}.")

        if user_choice == guardian_choice:
            print("It's a draw. The guardian lets you pass.")
        elif (user_choice == "rock" and guardian_choice == "scissors") or \
             (user_choice == "paper" and guardian_choice == "rock") or \
             (user_choice == "scissors" and guardian_choice == "paper"):
            print("You win! The guardian steps aside.")
        else:
            print("You lose! The guardian blocks your path.")
            return False

        # Second Choice
        choice2 = ask_for_choice("1. Take the glowing gem.\n2. Leave the gem and continue.\nEnter your choice (1/2): ")
        if choice2 == 1:
            print("The gem triggers a trap! You barely escape.")
            return False
        elif choice2 == 2:
            print("You find the hidden treasure. Victory!")
            return "won"
        else:
            return False
    
    elif choice1 == 2:
        print("You wander into a labyrinth and get lost.")
        return False
    else:
        return False


def run_game():
    """
    current_stage corresponds to index of stage_lst.
    current_stage is updated depending on if a stage is cleared in 3 attempts or not.
    Complete the game in 180 seconds.
    """
    current_stage = 0


    max_time = 180  # Maximum time for the game in seconds

    start_time = time.time()

    player = choose_character()
    print(f"\n[+] {player} Selected - Good Luck!")

    p_name = player.split("-")[0]


    while current_stage < len(stage_lst):
        # which challenge function to run based on current_stage
        if stage_lst[current_stage] == 'temple':
            stage_function = temple
        elif stage_lst[current_stage] == 'jungle':
            stage_function = jungle
        elif stage_lst[current_stage] == 'river':
            stage_function = river
        elif stage_lst[current_stage] == 'cave':
            stage_function = cave
            
        """
        try_stage return True if stage is completed successfully within 3 attempts
        try_stage return False if stage is not completed in 3 attempts.
        """        
        if try_stage(stage_function):
            current_stage += 1
        else:
            fail(p_name)
            break

        if time.time() - start_time > max_time:
            print("Time's up!")
            fail(p_name)
            break


    if current_stage == len(stage_lst):
        win(p_name)




welcome()
run_game()
```
